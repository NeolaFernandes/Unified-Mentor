Importing Dataset

```python
import pandas as pd

data = pd.read_excel("C:/Users/HP/OneDrive/Desktop/Unified Mentor/All states Census 2017 Population Estimates.xlsx")
data.head()
```

```python
# Display column names
print(data.columns)
```

```python
# Summary statistics
print(data.describe())
```

```python
# Data types and missing values
print(data.info())
print(data.isnull().sum())
```

```python
# Handling missing values, if any
data_cleaned = data.dropna()

# Verify the cleaned data
print(data_cleaned.info())
```

Population Growth Analysis

```python
# Add a new column for population growth from 2010 to 2017
data['population_growth_2010_2017'] = (
    (data['Population Estimate (as of July 1) - 2017'] - data['April 1, 2010 - Census']) / 
    data['April 1, 2010 - Census']) * 100

# Display the top 5 states with the highest growth
top_growth = data.sort_values(by='population_growth_2010_2017', ascending=False)[['Geography Name', 'population_growth_2010_2017']]
top_growth.head()
```

Top 5 Most and Least Populated States in 2017

```python
# Top 5 most populated states in 2017
top_5_populated = data.sort_values(by='Population Estimate (as of July 1) - 2017', ascending=False)[['Geography Name', 'Population Estimate (as of July 1) - 2017']].head()

# Top 5 least populated states in 2017
bottom_5_populated = data.sort_values(by='Population Estimate (as of July 1) - 2017', ascending=True)[['Geography Name', 'Population Estimate (as of July 1) - 2017']].head()

print(f"Top 5 most populated states in 2017:\n",top_5_populated,"\n")
print(f"Top 5 Least populated states in 2017:\n",bottom_5_populated)
```

Population Distribution Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(14, 8))
sns.histplot(data['Population Estimate (as of July 1) - 2017'], bins=20, kde=True, color='blue')
plt.title('Population Distribution Across States (2017)')
plt.xlabel('Population')
plt.ylabel('Number of States')
plt.grid(True)
plt.show()
```

State-wise Population Growth Visualization

```python
# Select a few states for visualization
selected_states = ['California', 'Texas', 'Florida', 'New York', 'Illinois']

# Filter data for the selected states
filtered_data = data[data['Geography Name'].isin(selected_states)]

# Plot the population estimates over time for these states
plt.figure(figsize=(14, 8))
for state in selected_states:
    plt.plot(range(2010, 2018), filtered_data[filtered_data['Geography Name'] == state].iloc[:, 4:12].values.flatten(), label=state)

plt.title('Population Growth (2010-2017) for Selected States')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.show()
```

Yearly Population Change Analysis

```python
# Calculate yearly population changes
data['population_change_2010_2011'] = data['Population Estimate (as of July 1) - 2011'] - data['Population Estimate (as of July 1) - 2010']
data['population_change_2011_2012'] = data['Population Estimate (as of July 1) - 2012'] - data['Population Estimate (as of July 1) - 2011']
data['population_change_2012_2013'] = data['Population Estimate (as of July 1) - 2013'] - data['Population Estimate (as of July 1) - 2012']
data['population_change_2013_2014'] = data['Population Estimate (as of July 1) - 2014'] - data['Population Estimate (as of July 1) - 2013']
data['population_change_2014_2015'] = data['Population Estimate (as of July 1) - 2015'] - data['Population Estimate (as of July 1) - 2014']
data['population_change_2015_2016'] = data['Population Estimate (as of July 1) - 2016'] - data['Population Estimate (as of July 1) - 2015']
data['population_change_2016_2017'] = data['Population Estimate (as of July 1) - 2017'] - data['Population Estimate (as of July 1) - 2016']

# Visualize the population change from 2010 to 2017 for selected states
plt.figure(figsize=(14, 8))
for state in selected_states:
    plt.plot(range(2011, 2018), data[data['Geography Name'] == state].iloc[:, -7:].values.flatten(), label=state)

plt.title('Yearly Population Change (2011-2017) for Selected States')
plt.xlabel('Year')
plt.ylabel('Population Change')
plt.legend()
plt.grid(True)
plt.show()
```

Population Growth Rate Comparison by Region

```python
# Dictionary to map states to their respective regions
regions = {
    'Northeast': ['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 
                  'New York', 'New Jersey', 'Pennsylvania'],
    'Midwest': ['Ohio', 'Michigan', 'Indiana', 'Illinois', 'Wisconsin', 'Minnesota', 'Iowa', 'Missouri',
                'North Dakota', 'South Dakota', 'Nebraska', 'Kansas'],
    'South': ['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 
              'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 
              'Arkansas', 'Louisiana', 'Oklahoma', 'Texas'],
    'West': ['Montana', 'Idaho', 'Wyoming', 'Colorado', 'New Mexico', 'Arizona', 'Utah', 'Nevada', 'Washington', 
             'Oregon', 'California', 'Alaska', 'Hawaii']
}

# Assign regions to the states
data['region'] = data['Geography Name'].apply(lambda x: next((region for region, states in regions.items() if x in states), 'Other'))

# Calculate average growth rate per region
region_growth = data.groupby('region')['population_growth_2010_2017'].mean()

# Visualize the regional growth rate comparison
plt.figure(figsize=(10, 6))
region_growth.plot(kind='bar', color='teal')
plt.title('Average Population Growth Rate (2010-2017) by Region')
plt.xlabel('Region')
plt.ylabel('Average Growth Rate (%)')
plt.grid(True)
plt.show()
```

Population Density Analysis

```python
land_area = {
    'California': 163696,
    'Texas': 268596,
    'Florida': 65758,
    'New York': 54555,
    'Illinois': 57914
}

# Calculate population density for 2017
data['land_area'] = data['Geography Name'].map(land_area)
data['population_density_2017'] = data['Population Estimate (as of July 1) - 2017'] / data['land_area']

# Visualize population density
plt.figure(figsize=(14, 8))
sns.barplot(x='Geography Name', y='population_density_2017', data=data[data['Geography Name'].isin(selected_states)], palette='magma')
plt.title('Population Density (2017) for Selected States')
plt.xlabel('State')
plt.ylabel('Population Density (people per square mile)')
plt.grid(True)
plt.show()
```

Correlation Analysis

```python
# Select population columns for correlation analysis
population_columns = data.columns[4:12]

# Calculate the correlation matrix
correlation_matrix = data[population_columns].corr()

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Population Estimates (2010-2017)')
plt.show()
```

Top Gainers and Losers Each Year

```python
# Find top gainers and losers for each year
years = range(2011, 2018)
for year in years:
    col_name = f'Population Estimate (as of July 1) - {year}'
    top_gainer = data.loc[data[col_name].idxmax()]['Geography Name']
    top_loser = data.loc[data[col_name].idxmin()]['Geography Name']
    max_increase = data[col_name].max()
    max_decrease = data[col_name].min()
    print(f"{year}: Top Gainer: {top_gainer} (+{max_increase}), Top Loser: {top_loser} ({max_decrease})")
```

Clustering States by Population Characteristics

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Select relevant columns for clustering
clustering_data = data.iloc[:, 4:12]

# Normalize the data
scaler = StandardScaler()
clustering_data_scaled = scaler.fit_transform(clustering_data)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=0)
data['Cluster'] = kmeans.fit_predict(clustering_data_scaled)

# Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Population Estimate (as of July 1) - 2010', y='Population Estimate (as of July 1) - 2017', hue='Cluster', palette='Set1')
plt.title('Clustering of States by Population Characteristics')
plt.xlabel('Population 2010')
plt.ylabel('Population 2017')
plt.grid(True)
plt.show()
```

Population Growth Trend Line

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Choose a state for trend analysis
state_name = 'Texas'
state_population = data[data['Geography Name'] == state_name].iloc[:, 4:12].values.flatten()

# Prepare data for linear regression
X = np.array(range(2010, 2018)).reshape(-1, 1)
y = state_population

# Fit linear regression model
model = LinearRegression()
model.fit(X, y)
trend_line = model.predict(X)

# Plot the trend line
plt.figure(figsize=(10, 6))
plt.plot(X.flatten(), y, label=f'Population of {state_name}', marker='o')
plt.plot(X.flatten(), trend_line, label='Trend Line', linestyle='--')
plt.title(f'Population Growth Trend for {state_name}')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.show()
```
