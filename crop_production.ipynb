```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:/Users/HP/OneDrive/Desktop/Unified Mentor/Crop Production data.csv")
data.head()
```

Distribution of Crop Production by States

```python
import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='State_Name', y='Production', ci=None)
plt.xticks(rotation=90)
plt.title('Total Crop Production by State')
plt.show()
```

Production Trends Over the Years for Different Crops

```python
plt.figure(figsize=(14, 8))
sns.lineplot(data=data, x='Crop_Year', y='Production', hue='Crop')
plt.title('Yearly Production Trends for Different Crops')
plt.legend(loc='upper right')
plt.show()
```

Area vs Production Scatter Plot (Colored by Crop Type)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Area', y='Production', hue='Crop', palette='Set2')
plt.title('Area vs Production for Different Crops')
plt.show()
```

Production Distribution by Crop Type

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Crop', y='Production', palette='Set1')
plt.xticks(rotation=90)
plt.title('Production Distribution by Crop Type')
plt.show()
```

Pairplot of Selected Features to Analyze Relationships

```python
sns.pairplot(data[['Area', 'Production', 'Season']], diag_kind='kde')
plt.suptitle('Pairplot of Key Features', y=1.02)
plt.show()
```

Correlation Matrix with Focused Variables

```python
corr_matrix = data[['Area', 'Production','Crop_Year']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Key Features')
plt.show()
```
